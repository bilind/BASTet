```python
from pylab import *
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import time
import sys
sys.path.append('/Users/bpb/Data/programming/OpenMSI/CodeForgeRepo/openmsi-tk/') #add path to your openmsi-tk location
from omsi.dataformat.omsi_file import *
from omsi.analysis.multivariate_stats.omsi_cx import *
```

## We'll use the openmsi file api to set a pointer to the peak cube.  Currently it is stored in "d"

```python
#Load the input data
inputFile = '/Users/bpb/20120711_Brain (2).h5'
f = omsi_file( inputFile , 'a' )
e = f.get_exp(0)
a = e.get_analysis(0)
d = a['peak_cube']
mz = a['peak_mz']
```

## Do a full SVD and keep the "rank" elements of the eigenvalue matrices.  The rows and columns of U and V that are most informative are the ones we want to keep.

```python
msidata = d[:]
originalShape = msidata.shape
rank = 5
# objectiveDim = self['objectiveDim'][0]
numBins   = msidata.shape[-1]
numPixels = msidata.size / numBins #The last data dimension is assumed to contain the spectra
msidata = msidata.reshape(numPixels, numBins).transpose()
#Compute the CX decomposition
U, D, V = np.linalg.svd(msidata, full_matrices=False)
levIons = np.sum(U[:,:rank]**2,axis=1)
levPixels = np.sum(V[:rank,:]**2,axis=0)

# levScores = self.comp_lev_exact(msidata, rank, objectiveDim)
infPixelIndices = levPixels.argsort()[::-1]
infIonIndices = levIons.argsort()[::-1]
```

## Reshape the informative indices into actual coordinates (ie: X, Y)

```python
[r,c] = np.unravel_index(infPixelIndices, [originalShape[0],originalShape[1]])
```

## Plot an image showing the informative locations.

```python
sx = np.argsort(levIons)
sx = sx[::-1]
plt.imshow(d[:,:,sx[1]])

# put a red dot, size 40, at 2 locations:
plt.scatter(c[:100], r[:100], c='r', s=40)

plt.show()
```

## Classify the informative ions as "good" or "bad" and use this informative to make a mask

```python
badIons = [0, 3, 4, 10, 12, 14]
goodIons = [1, 2, 5, 6, 7, 8, 9, 11, 13]
goodImage = np.zeros((d.shape[0],d.shape[1]))
badImage = np.zeros((d.shape[0],d.shape[1]))
maskImage = np.zeros((d.shape[0],d.shape[1]))

# goodImage = np.zeros((4,3))
# print goodImage.shape
for idx in goodIons:
    goodImage = goodImage + d[:,:,sx[idx]]
goodImage = goodImage / len(goodIons)  

for idx in badIons:
    badImage = badImage + d[:,:,sx[idx]]
badImage = badImage / len(badIons)  

# = d[:,:,sx[0]]
# good = d[:,:,sx[1]]
# goodIdx = (badImage < 20 and goodImage > 30)

maskImage[goodImage>30] = 1
maskImage[badImage>1000] = 0
# 
imshow(maskImage)
show()
```

```python
#Execute the analysis
maskD = d[:]
for i in range(0,maskD.shape[2]):
    maskD[:,:,i] = maskD[:,:,i]*maskImage
```
