```python
#Imports
%matplotlib inline
import numpy as np
import mkl
import matplotlib.pyplot as plt

import sys
def isneeded(x):
    if x not in sys.path:
        sys.path.append(x)

isneeded('/Users/curt/openMSI_SVN/openmsi-tk/')
isneeded('/Users/curt/openMSI_localdata/')

from omsi.dataformat.mzml_file_CF import *
```

```python
#Reading the MS2 image file and building a list of datacubes
omsi_ms2 = mzml_file(basename="/Users/curt/openMSI_localdata/re-reimaging_TI.mzML")
```

```python
#Plotting TIC images for each scan type
f, axarr = plt.subplots(3, 1, figsize=(6, 9))
for ind in range(len(omsi_ms2.data)):
    axarr[ind].imshow(omsi_ms2.data[ind][:, :, :].sum(axis=2).T)  #total ion chromatogram images
    axarr[ind].set_title('TIC for scan type '+omsi_ms2.scan_types[ind])
```

```python
#Plotting mass spectra for each scan type at a random pixel xbest, ybest 
## xbest, ybest is the point where MS1 TIC happened to be highest
xbest, ybest = np.unravel_index(np.argmax(omsi_ms2.data[1][:, :, :].sum(axis=2)), [omsi_ms2.shape[1][0], omsi_ms2.shape[1][1]])

f, axarr = plt.subplots(3, 1, figsize = (8, 12))
for ind in range(len(omsi_ms2.data)):
    axarr[ind].plot(omsi_ms2.mz[ind], omsi_ms2.data[ind][xbest, ybest, :])  #spectra
    axarr[ind].set_title(str(omsi_ms2.scan_types[ind]+' at pixel %s, %s') % (xbest, ybest))
```
