```python
#loading brain data
f = omsi_file('/Users/curt/openMSI_localdata/20120711_Brain.h5' , 'r+' )
e = f.get_experiment(0) #All data is organized as experiments and this just gets the first one
d= e.get_msidata(0) #This gets the omsi_file_msidata object for the first raw dataset 
print d.shape
print e.get_analysis_identifiers()
```

```python
#convolve the x, y domains using ndi
gaussfilt = ndi.filters.gaussian_filter
sigmas = [0, 0.75, 1.5]
ns = len(sigmas)
nx, ny, nmz = d.shape

start = time.time()
dat = d[:]
stop = time.time()
print 'Took %s seconds to load data' % (stop-start)

im_conv = [np.empty_like(d) for sigma in sigmas]

start = time.time()
for ind, sigma in enumerate(sigmas):
    im_conv[ind] = gaussfilt(dat, sigma = [sigma, sigma, 0])
    print 'Took %s seconds for gaussian smoothing with sigma = %s' % (time.time()-start, sigma)
```

```python
#plot the results

# finding the 868.8 ion range 
print np.argmin(abs(d.mz-868.8))
mzs = range(45450,45500)

#do the plots
f, ax = plt.subplots(1, ns, figsize=(16,4))
for ind in range(ns):
    ax[ind].imshow(im_conv[ind][:, :, mzs].sum(axis=2))
```
