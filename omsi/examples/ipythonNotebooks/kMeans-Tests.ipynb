```python
#Load the input data
inputFile = '/Users/bpb/20120711_Brain (2).h5'
# f = omsi_file('/Users/bpb/20140415_KBL_Brain_for_LowDose_GoodOneMiddle_80ss.h5' , 'r' )
# inputFile = '/Users/bpb/fingerprints_merged.h5'
f = omsi_file( inputFile , 'a' )
e = f.get_exp(0)
r = e.get_msidata(0)
```

```python
peakCube = omsi_findpeaks_global(name_key='Global peak finding on raw data')
# myIons = np.asarray([844.304, 868.288, 852.303, 806.331, 828.312, 804.314])
peakCube.execute( msidata=r, mzdata=r.mz, integration_width=0.1, peakheight=2, smoothwidth=3, slwindow=100)
```

```python
Nx = peakCube['peak_cube'].shape[0]
Ny = peakCube['peak_cube'].shape[1]
Nz = peakCube['peak_cube'].shape[2]
temp = np.amax(peakCube['peak_cube'],2)
data = np.divide(peakCube['peak_cube'],temp[:,:,np.newaxis])
data = peakCube['peak_cube'].reshape(Nx*Ny,Nz)
```

```python
from scipy.cluster.vq import kmeans, vq
centers, _ = kmeans(data, 12, iter=20)
cluster, _ = vq(data, centers)
```

```python
fig = plt.figure(1, (14., 14.))
grid = ImageGrid(fig, 111, # similar to subplot(111)
                nrows_ncols = (3, 4), # creates 2x2 grid of axes
                axes_pad=0.1, # pad between axes in inch.
                )

for i in range(12):
    grid[i].imshow(cluster.reshape(Nx,Ny)==i,cmap='gray') # The AxesGrid object work as a list of axes.
show()
```
