```python
#loading brain data
f = omsi_file('/Users/curt/openMSI_localdata/20120711_Brain.h5' , 'r+' )
e = f.get_experiment(0) #All data is organized as experiments and this just gets the first one
d= e.get_msidata(0) #This gets the omsi_file_msidata object for the first raw dataset 
print d.shape
```

```python
#rebin all of brain data
## manually defining new m/z axis
minmz = min(d.mz)
maxmz = max(d.mz)
nmz   = 1e3

mz_edges = np.linspace(minmz, maxmz, num=nmz+1)  #bin edges
mz_centers = np.diff(mz_edges)/2 + mz_edges[:-1:]  #bin centers found as the average of bin edges

mzrebin = omsi_mz_rebin(name_key='rebinning to 1e2 equally spaced points on m/z axis')

dnew, mz_centers = mzrebin.execute(msidata=d, mzdata=d.mz, new_mzdata=mz_centers, sigma=0, truncate=3.5)

a = e.create_analysis(mzrebin)
```

```python
#compare downsampled and upsampled data
f, ax = plt.subplots(1,2, figsize = (12,4))
ax[0].imshow(np.sum(d[:,:,(d.mz>867)&(d.mz<869)],axis=2))
ax[1].imshow(np.sum(dnew[:,:,(mz_centers>=850)&(mz_centers<890)],axis=2))


#plt.imshow(np.sum(d[:,:,(d.mz>867)&(d.mz<869)],axis=2))
```
