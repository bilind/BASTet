```python
#Better approach: do the mapping of old bins b (shape (I,)) to new bins n (shape (J,)) only once,  
#     forming a vector of increasing integer indices p (shape (I,), where p[i]==j if bin b[i] goes to bin[j]

#can histogramming be sped up by doing unweighted histogramming once and then just doing a faster vector multiply for each pixel?

# a new m/z axis with M points
M = 3e5
minmz = min(d.mz)
maxmz = max(d.mz)
nmz   = M  

mz_edges = np.linspace(minmz, maxmz, num=nmz+1)  #bin edges
mz_centers = np.diff(mz_edges)/2 + mz_edges[:-1:]  #bin centers found as the average of bin edges

#finding mapping between new bin centers and old bin centers, which only has to be done once for all pixels
q = np.searchsorted(d.mz, mz_centers)  # same shape as mz_centers, q[j] is index of dscan at which to stop cumsum for bin j
qAll = np.append(q, d.mz.shape[0])     # need to add the last bin which contains the data from the final bin edge to the end

#a single scan
dd = d[40, 40, :]

#a good gaussian filter size
filter_sigma = max(np.min(np.diff(d.mz)) / np.min(np.diff(mz_centers)),1)

#doing the rebinning on the single scan
zero = np.array(0, dtype=dd.dtype)
ddc = np.concatenate(([zero, ], dd.cumsum()))
H = np.diff(ddc[qAll])
Hnew = ndi.filters.gaussian_filter(H, sigma=filter_sigma )

f, ax = plt.subplots(1, 2, figsize = (20,6))
ax[0].plot(mz_centers, Hnew)
ax[0].set_xlim(865,875)

#doing rebinning on all pixes:
nx, ny, nold = d.shape
H = np.zeros((nx, ny, nmz))
start = time.time()
for ix in range(nx):
    for iy in range(ny):
        ddc = np.concatenate(([zero, ], d[ix, iy, :].cumsum()))
        Hscan = np.diff(ddc[qAll])
        H[ix, iy, :] = ndi.filters.gaussian_filter(Hscan, sigma=filter_sigma)
        if ix%10 == 0 and ix==iy:
            print ix, time.time() - start

stop = time.time()
print stop - start
ax[1].imshow(H[:, :, np.argmin(abs(mz_centers-868.8))])
```
