```python
from pylab import *
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import time
from omsi.dataformat.omsi_file import *
from omsi.analysis.multivariate_stats.omsi_cx import *
```

```python
#Load the input data
inputFile = '/Users/bpb/20120711_Brain (2).h5'
f = omsi_file( inputFile , 'a' )
e = f.get_exp(0)
a = e.get_analysis(0)
d = a['peak_cube']
```

```python
#Execute the analysis
ocxRaw = omsi_cx(nameKey='CX on Raw Data')
ocxRaw.execute( msidata=d , rank=6, objectiveDim=omsi_cx.dimension_index['imageDim'] )
```

```python
#Save the analysis to file
# e.create_analysis( ocx )
# f.close_file()
```

```python
sx = np.argsort(ocxRaw['levScores'])[::-1]
fig = plt.figure(1, (12., 5.))
subplot(121)
plot(a['peak_mz'],ocxRaw['levScores'])
subplot(122)
plot(ocxRaw['levScores'][sx])
show()
fig.savefig('/Users/bpb/testFig.eps')
```

```python
# imshow(d[:,:,0:3])
fig = plt.figure(1, (14., 14.))
grid = ImageGrid(fig, 111, # similar to subplot(111)
                nrows_ncols = (3, 5), # creates 2x2 grid of axes
                axes_pad=0.1, # pad between axes in inch.
                )

for i in range(15):
    grid[i].imshow(d[:,:,sx[i]]) # The AxesGrid object work as a list of axes.
show()
fig.savefig('/Users/bpb/preMask.eps')
```

```python
badIons = [0, 3, 4, 11, 15]
goodIons = [1, 2, 5, 6, 7, 8, 9, 10, 12, 13, 14]
goodImage = np.zeros((d.shape[0],d.shape[1]))
badImage = np.zeros((d.shape[0],d.shape[1]))
maskImage = np.zeros((d.shape[0],d.shape[1]))

# goodImage = np.zeros((4,3))
# print goodImage.shape
for idx in goodIons:
    goodImage = goodImage + d[:,:,sx[idx]]
goodImage = goodImage / len(goodIons)  

for idx in badIons:
    badImage = badImage + d[:,:,sx[idx]]
badImage = badImage / len(badIons)  

# = d[:,:,sx[0]]
# good = d[:,:,sx[1]]
# goodIdx = (badImage < 20 and goodImage > 30)

maskImage[goodImage>30] = 1
maskImage[badImage>1000] = 0
# 
imshow(maskImage)
show()
```

```python
#Execute the analysis
ocx = omsi_cx(nameKey='testCX')
maskD = np.zeros(d.shape[0],d.shape[1])
for i in range(0,maskD.shape[2]):
    maskD[:,:,i] = maskD[:,:,i]*maskImage
ocx.execute( msidata=maskD , rank=6, objectiveDim=omsi_cx.dimension_index['imageDim'] )
```
